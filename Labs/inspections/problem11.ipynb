```python
def deBrujin(seq, k):
    '''
    Creates a De Brujin graph represented as a ajecency list based on a given sequence and length of kmers
    :param seq: String representing 
    :param k:
    :return:
    '''
    brujinDict = {}
    for i in range(len(seq)-k):
        if seq[i:i+k-1] in brujinDict.keys():
            brujinDict[seq[i:i+k-1]].append(seq[i+1:i+k])
        else:
            brujinDict[seq[i:i+k-1]] = [seq[i+1:i+k]]

    return brujinDict

def main(fName):
    '''
    Reads input file and runs deBrujin on the data
    '''
    if fName == '':
        return
    else:
        with open(fName) as inFile:
            lines = inFile.readlines()
            k = int(lines[0].strip()) # first line is the value of k
            seq = lines[1].strip() # second line is the sequence
    dict = deBrujin(seq, k)
    toPrint = [] # used to organise output
    for key in dict.keys():
        new = '{0} -> '.format(key)
        for value in sorted(dict[key]):
            new += '{0},'.format(value)
        toPrint.append(new[:-1])

    toPrint.sort()
    for item in toPrint:
        print(item)
if __name__ == '__main__':
    main(fName='')
```
