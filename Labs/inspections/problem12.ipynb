```python
def deBrujin(kmers):
    deBrujinDict = {}

    for kmer in kmers:
        if kmer[:-1] in deBrujinDict:
            deBrujinDict[kmer[:-1]].append(kmer[1:])
        else:
            deBrujinDict[kmer[:-1]] = [kmer[1:]]

    return deBrujinDict
def main(fName):
    kmers = []
    if fName == '':
        return
    else:
        with open(fName) as inFile:
            for line in inFile:
                kmers.append(line.strip())

    DBdict = deBrujin(kmers)
    toPrint = []
    for key in DBdict.keys():
        new = '{0} -> '.format(key)
        for value in sorted(DBdict[key]):
            new += '{0},'.format(value)
        toPrint.append(new[:-1])

    toPrint.sort()
    for item in toPrint:
        print(item)
if __name__ == '__main__':
    main(fName='')
```
