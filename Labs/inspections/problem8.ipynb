```python
def genKmers(Seq, k):
    '''

    :param Seq: A string representing a DNA sequence
    :param k: the length of the k-mers, int
    :return: all k-mers in Seq as a list of strings
    '''
    kmers = []
    for i in range(len(Seq)-k +1): # iterates through sequence and finds all kmers of lenght k
        kmers.append(Seq[i:i+k])

    kmers.sort()
    return kmers

def main(fName):
    '''
    Opens input file and runs genKmers on the given data

    '''
    if fName == '':
        return
    with open(fName) as inFile: # opens file
        lines = inFile.readlines()
        k = int(lines[0].strip()) # first line is lenght of kmer
        seq = lines[1].strip() # next line is the sequence
    kmers = genKmers(seq,k)
    for kmer in kmers:
        print(kmer)

if __name__ == '__main__':
    main(fName='')
```

```python
main('rosalind_ba3a.txt')
```
