```python
import sys
def genSeq(kmers):
    '''
    A sequence of k-mers Pattern1, ... , Patternn such that the last k - 1 symbols of Patterni are equal to the first k - 1 symbols of Patterni+1 for i from 1 to n-1.
    :param kmers: list of strings, represnting individual kmers
    :return: a string represented the whole sequnce
    '''

    seq = kmers[0]
    for i in range(1,len(kmers)): # iterates through the kmers and adds the last character in each kmer to the growing seuqnce
        seq += kmers[i][-1]

    return seq
def main(fName):
    '''
    Reads input file and runs genSeq on the data
    returns the sequnce created from the input kmers
    '''
    kmers = []
    if fName == '':
        return ''
    else:
        with open(fName) as file:
            for line in file:
                kmers.append(line.strip())


    return genSeq(kmers)
if __name__ == '__main__':
    main(fName='')
```
